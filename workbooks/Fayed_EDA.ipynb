## Listing Column Names of Loaded DataFrames

```python
print("\n--- Columns for yelp_academic_dataset_business_df ---")
if 'yelp_academic_dataset_business_df' in globals():
    for col in yelp_academic_dataset_business_df.columns:
        print(col)
else:
    print("yelp_academic_dataset_business_df not found.")

print("\n--- Columns for yelp_academic_dataset_review_df ---")
if 'yelp_academic_dataset_review_df' in globals():
    for col in yelp_academic_dataset_review_df.columns:
        print(col)
else:
    print("yelp_academic_dataset_review_df not found.")
```

## Top 10 Categories in `yelp_academic_dataset_business_df`

```python
import collections

if 'yelp_academic_dataset_business_df' in globals():
    # Drop rows where 'categories' is NaN to avoid errors during splitting
    categories_series = yelp_academic_dataset_business_df['categories'].dropna()

    # Split the categories string into individual categories and flatten the list
    all_categories = []
    for category_list_str in categories_series:
        # Categories are typically comma-separated, sometimes with leading/trailing spaces
        categories_in_row = [cat.strip() for cat in category_list_str.split(',')]
        all_categories.extend(categories_in_row)

    # Count the occurrences of each category
    category_counts = collections.Counter(all_categories)

    print("\n--- Top 10 Most Frequent Categories ---")
    for category, count in category_counts.most_common(10):
        print(f"'{category}': {count}")
else:
    print("yelp_academic_dataset_business_df not found. Please ensure it is loaded.")
```

## Top States for Restaurants

```python
if 'yelp_academic_dataset_business_df' in globals():
    # Filter for businesses that have 'Restaurants' in their categories
    restaurants_df = yelp_academic_dataset_business_df[
        yelp_academic_dataset_business_df['categories'].str.contains('Restaurants', na=False)
    ]

    # Count the occurrences of each state for these restaurants
    top_restaurant_states = restaurants_df['state'].value_counts().head(20)

    print("\n--- Top 20 States with the Most Restaurants ---")
    display(top_restaurant_states)
else:
    print("yelp_academic_dataset_business_df not found. Please ensure it is loaded.")
```

## Restaurant Open/Closed Status

```python
if 'restaurants_df' in globals():
    # Count the number of open and closed restaurants
    open_closed_counts = restaurants_df['is_open'].value_counts()

    print("\n--- Open/Closed Status of Restaurants ---")
    display(open_closed_counts)
else:
    print("restaurants_df not found. Please ensure it is loaded and filtered.")
```

## Merging Business and Review Data for Average Reviews

```python
if 'yelp_academic_dataset_business_df' in globals() and 'yelp_academic_dataset_review_df' in globals():
    # Select necessary columns from business_df including 'review_count', 'stars' (business rating), and 'is_open'
    business_info_df = yelp_academic_dataset_business_df[['business_id', 'name', 'city', 'state', 'categories', 'stars', 'review_count', 'is_open']].copy()

    # Merge the two dataframes on 'business_id', including 'useful', 'funny', 'cool', and 'date' from review_df
    merged_df = pd.merge(business_info_df, yelp_academic_dataset_review_df[['business_id', 'stars', 'useful', 'funny', 'cool', 'date']],
                         on='business_id', how='left', suffixes=('_business', '_review'))

    print("Merged DataFrame created. Displaying head:")
    display(merged_df.head())

    print("\nNow, let's identify columns that can be used for averaging. The 'stars_review' column represents the individual review rating.")
    print("The 'stars_business' column represents the overall business rating provided in the business dataset.")
    print("Columns 'useful', 'funny', 'cool', and 'date' are also now available per review, and 'is_open' status is included.")

    # Calculate the average review stars for each business using 'stars_review'
    average_review_stars = merged_df.groupby(['business_id', 'name'])['stars_review'].mean().reset_index()
    average_review_stars = average_review_stars.rename(columns={'stars_review': 'average_review_stars'})

    # Merge 'is_open' status back to the average_review_stars DataFrame
    average_review_stars = pd.merge(average_review_stars, business_info_df[['business_id', 'is_open']], on='business_id', how='left').drop_duplicates()

    print("\nAverage review stars per restaurant (top 10), including 'is_open' status:")
    display(average_review_stars.sort_values(by='average_review_stars', ascending=False).head(10))

else:
    print("One or both of the required Dataframes (yelp_academic_dataset_business_df, yelp_academic_dataset_review_df) were not found.")
```

## Analyzing Reviews by Date Range

```python
# Ensure 'date' column is in datetime format
if 'merged_df' in globals() and 'date' in merged_df.columns:
    merged_df['date'] = pd.to_datetime(merged_df['date'])
    print("\n'date' column converted to datetime format.")
    print(f"Date range of reviews: {merged_df['date'].min()} to {merged_df['date'].max()}")
else:
    print("merged_df not found or 'date' column missing.")
```

```python
if 'merged_df' in globals():
    # Define the date range (example: adjust these dates as needed)
    start_date = '2019-01-01'
    end_date = '2019-12-31'

    # Filter the merged DataFrame by date range
    filtered_merged_df = merged_df[(merged_df['date'] >= start_date) & (merged_df['date'] <= end_date)]

    print(f"\nFiltered reviews from {start_date} to {end_date}. Displaying head:")
    display(filtered_merged_df.head())

    # Recalculate average review stars for each business within the date range
    average_review_stars_date_range = filtered_merged_df.groupby(['business_id', 'name'])['stars_review'].mean().reset_index()
    average_review_stars_date_range = average_review_stars_date_range.rename(columns={'stars_review': 'average_review_stars'})

    # Merge 'is_open' status back to the average_review_stars_date_range DataFrame
    # Use business_info_df which has the original is_open status for each business
    if 'business_info_df' in globals():
        average_review_stars_date_range = pd.merge(average_review_stars_date_range, business_info_df[['business_id', 'is_open']], on='business_id', how='left').drop_duplicates()

    # Add start_date and end_date columns to the DataFrame
    average_review_stars_date_range['start_date'] = start_date
    average_review_stars_date_range['end_date'] = end_date

    print(f"\nAverage review stars per restaurant (top 10) for reviews between {start_date} and {end_date}, including 'is_open' status, start_date and end_date:")
    display(average_review_stars_date_range.sort_values(by='average_review_stars', ascending=False).head(10))

else:
    print("merged_df not found. Please ensure it is loaded.")
```
